```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth"
page=requests.get(url)
soup=BeautifulSoup(page.content,"lxml")
soup
```
